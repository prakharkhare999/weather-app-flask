# seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Create a sample DataFrame with 100 rows
data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),         # Categorical variable with 3 categories
    'Value1': np.random.randn(100) * 10 + 50,                          # Normally distributed numeric variable
    'Value2': np.random.rand(100) * 100,                               # Uniformly distributed numeric variable
    'Group': np.random.choice(['X', 'Y'], size=100)                    # Another categorical variable for grouping
}

df = pd.DataFrame(data)
print("Sample DataFrame:")
print(df.head())

# Set seaborn style
sns.set(style="whitegrid")

# Example 1: Boxplot of Value1 by Category
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value1', data=df)
plt.title('Boxplot of Value1 by Category')
plt.show()

# Example 2: Scatterplot of Value1 vs. Value2 with Group hue and Category style
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Group', style='Category', data=df)
plt.title('Scatterplot of Value1 vs Value2')
plt.show()

# Example 3: Violin plot of Value2 by Group
plt.figure(figsize=(8, 6))
sns.violinplot(x='Group', y='Value2', data=df)
plt.title('Violin Plot of Value2 by Group')
plt.show()
```
